## ðŸ§  Why I added it:
In many modern e-commerce sites (like Flipkart), products are nested inside repeated blocks â€” often <div>s with a unique class name that wraps all product info (name, price, rating, etc.).

I guessed tUxRFH might be one of those individual product containers, because Flipkart frequently changes their HTML structure and uses random-looking class names for each product block.

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests as r
import numpy as np
import time

product_name =[]
product_price = []
product_description = []
product_rating = []

for page in range(1,50):
    HEADERS = ({
        'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Version/4.0 Chrome/130.0.6723.58 Safari/537.36 (AirWatch Browser v21.09.0.9)"',
        'Accept-Language':'en-Us, en;q=0.5'})
    URL = f'https://www.flipkart.com/search?q=phoe&sort=price_desc&page={page}'
    webpage = r.get(URL, headers=HEADERS)
    # link = soup.find("a",class_="_9QVEpD").get("href")
    # clink = "https://www.flipkart.com"+link
    soup = BeautifulSoup(webpage.text, "lxml")
    box = soup.find_all("div", class_="DOjaWF gdgoEp")

    for containers in box:
        prod = containers.find_all("div",class_="tUxRFH")
        for product in prod:
                 try:
                        name = product.find("div",class_="KzDlHZ").text
                 except:
                        name= None
                 try:
                        desc = product.find("ul",class_="G4BRas").text
                 except:
                        desc = None
                 try:
                        price = product.find("div",class_="Nx9bqj _4b5DiR").text
                 except:
                        price = None
                 try:
                        rating = product.find("div", class_="XQDdHH").text
                 except:
                        rating = None
        
                 product_name.append(name)
                 product_description.append(desc)
                 product_price.append(price)
                 product_rating.append(rating)
            
        time.sleep(1)  # pause between requests

print(len(product_name), "names")
print(len(product_price), "prices")
print(len(product_description), "descriptions")
print(len(product_rating), "ratings")

df = pd.DataFrame({
    "Names":product_name,
    "Description":product_description,
    "Prices":product_price,
    "Ratings":product_rating
})
print(df)

df["Prices"]=df["Prices"].str.replace("â‚¹","").str.replace(",","").str.strip()
df["Prices"]=pd.to_numeric(df["Prices"],errors="coerce").fillna(0).astype(int)
# print(df["Prices"])
# Clean and convert Rating column
df["Ratings"] = pd.to_numeric(df["Ratings"], errors="coerce").astype(float)
df.to_csv("flipkart_product1.csv", index=False)
print("sucessfully completed 'flipkart_products1.csv'")
```
